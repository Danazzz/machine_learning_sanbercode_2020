Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# membuat component figure dan axis
fig, ax = plt.subplots()

# memberikan data kedalam axis
ax.plot(x, cos, color='#FF0000',label='Cosx')
ax.plot(x, cos3, linestyle='--', color='#00FF00',label='3 Cosx')
ax.plot(x, sin, linestyle='-.', color='#0000FF',label='sin{X-90}')
ax.plot(x, sin2, marker='x', linestyle='', color='#FFFF00',label='2 Sinx')

# mengubah axis label dan title
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')

#membuat legenda
ax.legend(loc='center',framealpha=0.4)

#tampilkan
fig.set_size_inches(10,6.8)
plt.show()
```
