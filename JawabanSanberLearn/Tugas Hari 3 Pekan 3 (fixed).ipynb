```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#membuat figure dan multiple axis
fig, ax = plt.subplots(figsize=(17,8), ncols=3, sharex=True, sharey=True)

#figure ini adalah 1D array yang memiliki 3 kolom
ax[0].plot(x, tan ,color='#0000FF',label='tanX')
ax[1].plot(x, sin, color='#FF0000',label='sinX')
ax[2].plot(x, cos, color='#00FF00',label='cosX')

#memberi title pada tiap axis
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')

#membuat legenda
fig.legend(loc='upper center')

#tampilkan
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#membuat figure dan axis
fig, ax = plt.subplots(figsize=(12.8,8))
ax.plot(x,y)
ax.set_title('Grafik CosX')

#membuat axis yang akan di nested kedalam axis
cos2= ax.inset_axes([-0.1,-0.1,0.3,0.3])
cos2.plot(x,y2,color='#FF0000')
cos2.plot(x,y,color='#0000FF')
cos2.set_title('Grafik CosX^2')

cos3= ax.inset_axes([0.9,-0.1,0.3,0.3])
cos3.plot(x,y3,color='#008800')
cos3.plot(x,y,color='#0000FF')
cos3.set_title('Grafik CosX^3')

cos4= ax.inset_axes([0.9,0.9,0.3,0.3])
cos4.plot(x,y4,color='#010101')
cos4.plot(x,y,color='#0000FF')
cos4.set_title('Grafik CosX^4')

cos5= ax.inset_axes([-0.1,0.9,0.3,0.3])
cos5.plot(x,y5,color='#ffff00')
cos5.plot(x,y,color='#0000FF')
cos5.set_title('Grafik CosX^5')

#tampilkan
plt.show()
```
